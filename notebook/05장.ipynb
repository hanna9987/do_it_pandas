```python
ebola_long = ebola.melt(['Date', 'Day'])
```

```python
ebola_long.variable
# 이거는 시지르임. 스플릿 불가능.
```

```python
ebola_long.variable.str
# 시리즈를 str 문자열로 바꿔서 매서드 가능하게 함.
```

```python
ebola_long.variable.str.split('_')
```

```python
ebola_long.variable.str.split('_')[0]
```

```python
ebola_long.variable.str.split('_').str.get(0)
```

```python
ebola_long.variable.str.split('_').str.get(1)
```

```python
ebola_long['status'] = ebola_long.variable.str.split('_').str.get(0)
ebola_long['country'] = ebola_long.variable.str.split('_').str.get(1)
```

```python
ebola_long
```

```python
ebola_long.variable.str.split('_',expand=True)
# expand 는 이거 한번해 분할해줌.
```

```python
ebola_long['status'] = ebola_long.variable.str.split('_',expand=True)[0]
ebola_long['country'] = ebola_long.variable.str.split('_',expand=True)[1]
```

```python
ebola_long
```

```python
ebola_long[['status','country']] = ebola_long.variable.str.split('_',expand=True)
# 이렇게 한번에도 가능. 다중 분할 기능. 이렇게 단순한거는 가능하지만 대부분 하나씩 해야함.
```

```python
w_melt.pivot_table(index=['id', 'year', 'month', 'day'],
                  columns='element',
                   values='temp')
```

```python
w_tiny = w_melt.pivot_table(index=['id', 'year', 'month', 'day'],
                  columns='element',
                   values='temp')
```
